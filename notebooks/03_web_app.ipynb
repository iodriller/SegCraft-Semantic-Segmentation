# SegCraft web app

The optional FastAPI app gives you a small browser UI for video upload, YouTube input, progress, and output downloads.

## 1) Install the app extras

Run this in a terminal from the repository root:

```python
# pip install -e ".[web]"
```

## 2) Start the server

```python
# segcraft-web
```

Open `http://127.0.0.1:8000`. Upload a local video or paste a YouTube URL. The app writes jobs under `outputs/web/jobs/` and exposes downloads for the generated videos and summary JSON.

For the Cityscapes demo model from the repo checkout, set the preset field to `configs/presets/cityscapes_video.yaml`. For a wheel install without the repo configs, leave the preset blank and provide your own config path when needed.
