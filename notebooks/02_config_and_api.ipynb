# SegCraft config and API tour

This notebook uses the lightweight parts of SegCraft: config loading, typed sections, dataset file pairing, and model-spec resolution. It does not require PyTorch.

## 1) Load the base config

```python
from segcraft import load_config, load_config_object

config = load_config("configs/base.yaml")
typed = load_config_object("configs/base.yaml")

typed.task, typed.model, typed.data.image_size
```

## 2) Resolve a model spec without importing heavy ML libraries

```python
from segcraft.models import build_model

build_model(typed.model, typed.task)
```

## 3) Pair image and mask files by filename stem

```python
from pathlib import Path
from segcraft.data import pair_image_masks

root = Path("data/notebook_demo")
images = root / "images"
masks = root / "masks"
images.mkdir(parents=True, exist_ok=True)
masks.mkdir(parents=True, exist_ok=True)

(images / "frame_001.jpg").write_text("demo image placeholder", encoding="utf-8")
(masks / "frame_001.png").write_text("demo mask placeholder", encoding="utf-8")

[pair.to_dict() for pair in pair_image_masks(images, masks)]
```

For real training or prediction, install the relevant optional extras such as `pip install -e ".[torch]"`, then point `configs/local.yaml` at real images, masks, or a video file.
